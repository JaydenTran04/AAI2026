```python
import random

random.seed(42)

missions = []
answers = []

for i in range(10):
    fuel_start = random.randint(1, 10)
    fuel_change = random.randint(1, 10)
    operation = random.choice(['+', '-'])  # refuel or burn fuel

    if operation == '+':
        result = fuel_start + fuel_change
        story = "Refuel"
        symbol = "+"
    else:
        result = fuel_start - fuel_change
        story = "Fuel Burn"
        symbol = "-"

    missions.append(
        f"{i+1}. Space Cadet Mission: {story} | "
        f"Fuel: {fuel_start} {symbol} {fuel_change} = ____ units"
    )
    answers.append(f"{i+1}. {result} units")

print("🚀 Space Cadet Training Missions:\n")
for m in missions:
    print(m)

print("\n" + "=" * 35)
print("✅ Answer Key:\n")
for a in answers:
    print(a)
```
