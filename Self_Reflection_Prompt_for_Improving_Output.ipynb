```python
import re
from typing import List, Tuple

class QuestCoach:
    def __init__(self):
        self.max_words = 120

    def count_words(self, text: str) -> int:
        """Count words in text"""
        return len(text.split())

    def identify_issues(self, draft: str) -> List[Tuple[str, str, str]]:
        """
        Identify common quest-log issues
        Returns: List of (issue_type, explanation, why_it_matters)
        """
        issues = []

        # Check for passive voice
        passive_patterns = [r'\b(is|are|was|were|be|been|being)\s+\w+ed\b']
        for pattern in passive_patterns:
            if re.search(pattern, draft, re.IGNORECASE):
                issues.append((
                    "Passive Voice Detected",
                    "Found passive constructions (e.g., 'was done', 'is being')",
                    "Active voice makes your quest log clearer and more decisive"
                ))
                break

        # Check for wordiness
        wordy_phrases = {
            'in order to': 'to',
            'due to the fact that': 'because',
            'at this point in time': 'now',
            'a number of': 'several/many'
        }
        found_wordy = []
        for wordy, concise in wordy_phrases.items():
            if wordy in draft.lower():
                found_wordy.append(f"'{wordy}' → '{concise}'")

        if found_wordy:
            issues.append((
                "Wordiness",
                f"Replace: {', '.join(found_wordy)}",
                "Shorter instructions are easier to follow during a mission"
            ))

        # Check for vague language
        vague_words = ['things', 'stuff', 'very', 'really', 'somewhat', 'kind of', 'sort of']
        found_vague = [word for word in vague_words if word in draft.lower()]
        if found_vague:
            issues.append((
                "Vague Language",
                f"Found imprecise words: {', '.join(found_vague)}",
                "Specific details reduce confusion and prevent mistakes"
            ))

        # Check for multiple sentences without clear structure
        sentences = [s.strip() for s in re.split(r'[.!?]+', draft) if s.strip()]
        if len(sentences) > 5 and not any(word in draft.lower() for word in ['first', 'second', 'however', 'therefore', 'additionally']):
            issues.append((
                "Lack of Transitions",
                "Multiple sentences without connecting words",
                "Transitions help players understand the steps in order"
            ))

        # Check word count
        word_count = self.count_words(draft)
        if word_count > self.max_words:
            issues.append((
                "Exceeds Word Limit",
                f"Current: {word_count} words (limit: {self.max_words})",
                "Short quest logs keep the mission focused"
            ))

        # Check for objectives and next actions
        goal_keywords = ['goal', 'objective', 'aim', 'target', 'achieve']
        action_keywords = ['will', 'plan to', 'next step', 'action', 'schedule']

        has_goal = any(keyword in draft.lower() for keyword in goal_keywords)
        has_action = any(keyword in draft.lower() for keyword in action_keywords)

        if not has_goal:
            issues.append((
                "Missing Objective",
                "No clear mission objective stated",
                "Objectives tell the player what success looks like"
            ))

        if not has_action:
            issues.append((
                "Missing Next Move",
                "No specific next action or timing mentioned",
                "A next move ensures the mission progresses"
            ))

        return issues

    def generate_critique(self, draft: str) -> str:
        """Generate a critique of the quest log"""
        issues = self.identify_issues(draft)

        if not issues:
            return "✓ No major issues found! Consider adding a clear objective and next move."

        critique = "🗺️ **QUEST COACH REPORT**\n\n"

        for i, (issue_type, explanation, why_it_matters) in enumerate(issues, 1):
            critique += f"**{i}. {issue_type}**\n"
            critique += f"   - Issue: {explanation}\n"
            critique += f"   - Why it matters: {why_it_matters}\n\n"

        return critique

    def generate_improved_version(self, draft: str) -> str:
        """
        Generate an improved version with guidelines
        This is a template - in practice, you'd rewrite based on the actual draft
        """
        improvement = "\n✨ **IMPROVED QUEST LOG** (Template)\n\n"
        improvement += "[Rewrite the quest log to be:\n"
        improvement += "- ≤120 words\n"
        improvement += "- Active voice and concise\n"
        improvement += "- Include ONE clear objective (specific, measurable)\n"
        improvement += "- Include ONE next move (actionable, with timeline)\n"
        improvement += "- Clear structure with transitions]\n\n"
        improvement += "**Example Structure:**\n"
        improvement += "[Situation/Context in 2-3 sentences]\n\n"
        improvement += "**Objective:** [One specific, measurable objective]\n\n"
        improvement += "[1-2 sentences on approach/reasoning]\n\n"
        improvement += "**Next Move:** [One actionable step with timeline]"

        return improvement

    def analyze(self, draft: str):
        """Main analysis function"""
        print("=" * 60)
        print("QUEST COACH: LOG ANALYSIS")
        print("=" * 60)
        print(f"\n📜 **ORIGINAL QUEST LOG** ({self.count_words(draft)} words)\n")
        print(draft)
        print("\n" + "=" * 60 + "\n")

        # Generate critique
        critique = self.generate_critique(draft)
        print(critique)
        print("=" * 60)

        # Generate improved version guidance
        improved = self.generate_improved_version(draft)
        print(improved)
        print("\n" + "=" * 60)
        print("\n💡 Tip: Apply these changes to finalize your quest log!")

def main():
    """Main function for Google Colab"""
    print("╔══════════════════════════════════════════════════════════╗")
    print("║            QUEST COACH: QUEST LOG REVIEW TOOL            ║")
    print("╚══════════════════════════════════════════════════════════╝\n")

    # Get user input
    print("Paste your quest log below (press Enter twice when done):")
    print("-" * 60)

    lines = []
    empty_count = 0

    while True:
        try:
            line = input()
            if line == "":
                empty_count += 1
                if empty_count >= 2:
                    break
            else:
                empty_count = 0
                lines.append(line)
        except EOFError:
            break

    draft = "\n".join(lines).strip()

    if not draft:
        print("\n❌ No quest log provided. Please try again.")
        return

    # Analyze the draft
    coach = QuestCoach()
    coach.analyze(draft)

    # Option to analyze another quest log
    print("\n" + "=" * 60)
    print("Would you like to review another quest log? (yes/no)")
    response = input().strip().lower()

    if response in ['yes', 'y']:
        print("\n" * 2)
        main()
    else:
        print("\n✅ Thanks for using Quest Coach! Good luck on your adventure.")

# Run the program
if __name__ == "__main__":
    main()
```
